```python
from pathlib import Path
import subprocess

DATA_ROOT = Path('/kaggle/input/competitions/omni-sub')
SAMPLE_PATH = DATA_ROOT / 'sample_submission.csv'
CKPT_PATH = Path('runs/run_char_warmstart_overfit/best.pt')
beam_csv = Path('submission_beam.csv')
final_csv = Path('submission_beam_wordnorm.csv')

subprocess.run([
    'python', 'infer_beam_submission.py',
    '--ckpt', str(CKPT_PATH),
    '--data-root', str(DATA_ROOT),
    '--sample-submission', str(SAMPLE_PATH),
    '--output-csv', str(beam_csv),
    '--beam-size', '30', '--batch-size', '6', '--num-workers', '0'
], check=True)

subprocess.run([
    'python', 'apply_wordnorm_from_beam.py',
    '--input-csv', str(beam_csv),
    '--output-csv', str(final_csv),
], check=True)
print('done:', final_csv)
```
