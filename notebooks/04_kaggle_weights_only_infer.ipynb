# ???????????: Kaggle weights-only inference

```python
from pathlib import Path
import subprocess

INPUT_ROOT = Path('/kaggle/input')
sample_candidates = list(INPUT_ROOT.rglob('sample_submission.csv'))
ckpt_candidates = list(INPUT_ROOT.rglob('best.pt'))
script_path = Path('scripts/kaggle_infer_weights_only.py')
if not sample_candidates:
    raise RuntimeError('sample_submission.csv not found')
if not ckpt_candidates:
    raise RuntimeError('best.pt not found')
sample_path = sample_candidates[0]
data_root = sample_path.parent
ckpt_path = ckpt_candidates[0]
out_csv = Path('/kaggle/working/submission_final_beam_wordnorm.csv')
cmd = [
    'python', str(script_path),
    '--ckpt', str(ckpt_path),
    '--data-root', str(data_root),
    '--sample-submission', str(sample_path),
    '--output-csv', str(out_csv),
    '--beam-size', '30', '--batch-size', '6', '--num-workers', '2', '--apply-wordnorm'
]
subprocess.run(cmd, check=True)
print('saved:', out_csv)
```
